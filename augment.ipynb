```python
import random
```

```python
possible = [
      vidaugs.AddNoise(level=random.choice([15, 25, 35])),
      vidaugs.Brightness(level=random.choice([-0.4, -0.2, 0.2, 0.4])),
      vidaugs.ChangeVideoSpeed(factor=random.choice([0.1, 0.3, 1.1, 1.3])),
      vidaugs.Blur(sigma=random.choice([0.1, 0.3, 0.5])),
      vidaugs.Contrast(level=random.randint(-50, 50))
  ]
```

```python
def get_augmentation_techniques():
  number = random.randint(1, 3)
  number_of_techniques = random.sample([i for i in range(5)], k=number)
  return number_of_techniques
```

```python
def augment_data(path):
  output_path = "/content/B5_augmented" + path[8:-4] + "augmented.mp4"
  techniques = get_augmentation_techniques()
  aug = vidaugs.Compose(
      [possible[techniques[i]] for i in range(len(techniques))]
  )
  aug(path, output_path)
```

```python
for path in paths:
  augment_data('/content/' + path)
```
